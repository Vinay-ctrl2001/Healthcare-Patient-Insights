```python
# Convert Patient Admission Date to datetime
df['Patient Admission Date'] = pd.to_datetime(
    df['Patient Admission Date'], errors='coerce')
df
```

```python
# Convert Patient Admission Time to datetime
def parse_time_column(series):
    # Try multiple formats in order
    formats = ['%H:%M:%S', '%H:%M', '%I:%M %p']
    parsed = pd.Series([pd.NaT] * len(series), index=series.index)

    for fmt in formats:
        mask = parsed.isna()
        try:
            parsed[mask] = pd.to_datetime(series[mask], format=fmt, errors='coerce')
        except Exception:
            pass


   mask = parsed.isna()
    parsed[mask] = pd.to_datetime(series[mask], errors='coerce')

    return parsed.dt.time

df = pd.read_csv(r"C:\Users\Vinay\Downloads\healthcare_data.csv")  # reload fresh data
df['Patient Admission Time'] = parse_time_column(df['Patient Admission Time'])
df
```

```python
# Drop columns
df = df.drop(columns = "Merged")
df
```

```python
# Correcting values
df['Patient Gender'] = df['Patient Gender'].replace('Femaleemale','Female')
df
```

```python
# Adding new Age Group column
bins = [0, 18, 65, 100]
labels = ['Child', 'Adult', 'Senior']

# Create the 'Age Group' column
df['Age Group'] = pd.cut(df['Patient Age'], bins=bins, labels=labels, right=True)

# Move 'Age Group' to column index 5
col = df.pop('Age Group')
df.insert(5, 'Age Group', col)
df
```

```python
# Replacing none with NO referral
df['Department Referral'] = (
    df['Department Referral']
    .fillna('No Referral')   
    .astype(str)
    .str.lower()
    .str.strip()
    .replace('none', 'no referral') 
)
df
```

```python
# Changing column values to Yes No
df['Patient Admission Flag'] = df['Patient Admission Flag'].replace({'Admission':'Yes', 'Not Admission':'No'})
df
```

```python
# Filling NaN values
median_satisfaction_score = df['Patient Satisfaction Score'].median()
df['Patient Satisfaction Score'].fillna(median_satisfaction_score, inplace = True)
df
```

```python
# changing column name
df = df.rename(columns={'Patient Waittime': 'Patient Waittime mins'})
df
```
